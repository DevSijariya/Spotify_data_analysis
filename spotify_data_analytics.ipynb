Changing the index values as release date



```python
df_data.set_index("release_date",inplace=True)
df_data.index=pd.to_datetime(df_data.index)
df_data.head()
```

Now Checking the artist at any row using the index value

```python
# Before converting
df_data.duration_ms.head()
```

```python
df_data["duration"]=df_data["duration_ms"].apply(lambda x:round(x/1000))
df_data.drop("duration_ms",inplace=True,axis=1)
```

```python
# After converting
df_data.duration.head()
```

Now find the Correlation

```python
corr_df=df_data.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(15,7))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("Coreelation Heatmap between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

Number of songs per year


```python
df_data['dates']=df_data.index.get_level_values('release_date')
df_data.dates=pd.to_datetime(df_data.dates)
years=df_data.dates.dt.year
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per year")
```

```python
total_df=df_data.duration
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_df,ax=ax,errwidth=False).set(title="Year Vs Duration")
plt.xticks(rotation=90)
```

```python
total_df=df_data.duration
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=total_df,ax=ax).set(title="Year Vs Duration")
plt.xticks(rotation=60)
```

```python
plt.title("Durations of the Songs In Differnt Genres")
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y='genre',x="duration_ms",data=features_df)
plt.xlabel('Duration in Milisecond')
plt.ylabel('Genres')
```
